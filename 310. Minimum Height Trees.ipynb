### 310. Minimum Height Trees

```python
class Solution:
    def findMinHeightTrees(self, n, edges):
        # 建立一個空的圖形列表，每個節點都對應一個集合，用來儲存與該節點相鄰的其他節點
        graph = [set() for _ in range(n)]
        print(f"n = {n}, edges = {edges}")
        
        # 將邊的資訊加入到圖形中，每個節點的集合中都會包含與之相鄰的節點
        for u, v in edges:
            graph[u].add(v)
            graph[v].add(u)
            print("-" * 200)
            print(f"u = {u}, v = {v}")
            print(f"graph = {graph}")

        # 找出所有葉節點，即僅有一個鄰居節點的節點。初始時，這些葉節點被視為可能的最小高度樹的根節點。
        leaves = [x for x in range(n) if len(graph[x]) <= 1]
        prev_leaves = leaves
        
        # 進入循環，直到找到最小高度樹的根節點為止
        while leaves:
            new_leaves = []
            print("_" * 200)         
            print(f"prev_leaves = {prev_leaves}, leaves = {leaves}")
            
            # 遍歷leaves中的每個葉節點
            for leaf in leaves:
                print("-" * 200)
                print(f"leaf = {leaf}, graph[{leaf}] = {graph[leaf]}")            
                if not graph[leaf]: # 檢查該葉節點是否已經沒有鄰居節點，如果是，則它就是潛在的最小高度樹的根節點
                    return leaves
                
                neighbor = graph[leaf].pop() # 否則，取出該葉節點的唯一鄰居節點
                graph[neighbor].remove(leaf) # 從graph中刪除這個葉節點和neighbor之間的邊，以維護圖的結構
                print(f"neighbor = {neighbor}, graph = {graph}")

                print(f"(len(graph[neighbor]) = {len(graph[neighbor])}) == 1")
                if len(graph[neighbor]) == 1: # 如果鄰居節點也成為新的葉節點
                    new_leaves.append(neighbor) # 將其添加到新的葉節點列表
                    print(f"new_leaves = {new_leaves}")
                    
            prev_leaves, leaves = leaves, new_leaves # 更新前一輪和當前葉節點列表

        return prev_leaves  # 返回最小高度樹的根節點
```

```python
n = 4
edges = [[1,0],[1,2],[1,3]]  

n = 6
edges = [[3,0],[3,1],[3,2],[3,4],[5,4]]  

Solution().findMinHeightTrees(n, edges)
```
