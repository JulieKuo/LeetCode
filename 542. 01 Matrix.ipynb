### 542. 01 Matrix

```python
from collections import deque
class Solution():
    def updateMatrix(self, mat):
        print(f"mat = {mat}")
        if not mat or not mat[0]:
            return []

        m, n = len(mat), len(mat[0])
        print(f"m = {m}, n = {n}")

        queue = deque()
        max_value = m * n # 最大距離
        print(f"max_value = {max_value}")
        
        # 把所有為0的座標放進queue中，並將不為0的改為max_value
        for i in range(m):
            for j in range(n):
                if mat[i][j] == 0:
                    queue.append((i, j))
                else:
                    mat[i][j] = max_value
        print(f"queue = {queue}")
        print(f"mat = {mat}")
        
        directions = [(1, 0), (-1, 0), (0, -1), (0, 1)] # 上下左右
        while queue:
            row, col = queue.popleft() # 依序抓出queue的座標
            print(f"{'-' * 200}")
            print(f"row = {row}, col = {col}, mat[row][col] = {mat[row][col]}")
            for dr, dc in directions: # 依序往上下左右移動
                r, c = row + dr, col + dc # 更新座標
                if 0 <= r < m and 0 <= c < n: # 如果座標在範圍內       
                    print(f"- r = {r}, c = {c}, mat[r][c] = {mat[r][c]}")
                    if mat[r][c] == max_value: # 如果座標的值為max_value，代表尚未更新與0的最近距離
                        print(f"init: mat = {mat}")
                        queue.append((r, c)) # 加入queue，它周圍可能也有不為0的值
                        mat[r][c] = mat[row][col] + 1 # 中心點的值+1
                        print(f"new:  mat = {mat}")
                        print(f"queue = {queue}")
        
        return mat
```

```python
# mat = [
#     [0,0,0],
#     [0,1,0],
#     [0,0,0]
#     ]

mat = [
    [0,0,0],
    [0,1,0],
    [1,1,1]
    ]  
Solution().updateMatrix(mat)
```
