### 383. Ransom Note (Anonymous Notes)

#### Question
Your task this week is to help your shy friend to write an anonymous note to a classmate. You have a few pages from a magazine available, and the message that your friend wants to compose. The plan is to cut words from the magazine and paste them on a sheet of paper in the right order to compose the final message.

Before starting this tedious activity, you would like to make sure that the pages contain all the words you will need. Your task is to prepare a function that takes as input two strings:

- the first one is the text from the magazine
- the second one is the message that your friend wants to send

Your function should return True if the message can be composed with the words in the magazine, False otherwise.

For simplicity, you can assume that strings have only words separated by spaces, without punctuation.

> [!NOTE]: Include punctuation as a followup.

#### Examples
If the students asks for an example, offer these:

```
"Our 2 bears live in a den 嗎", "In our den 嗎" -> True
```

#### Hash

**時間複雜度: $O(n + m)$**  
**空間複雜度: $O(n)$**

- 效率最高

````python
class Solution:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        # 初始化雜湊表，用來儲存字母出現次數
        counts = {} # space: O(n)

        # 遍歷第一個字串 magazine
        for char in magazine: # time: O(n)
            # 判斷是否為英文字母（忽略標點與空格）
            if char.isalnum():
                # 統一轉為小寫
                char = char.lower()
                # 記錄字母出現次數：若不存在則設為 0，然後 +1
                counts[char] = counts.get(char, 0) + 1

        # 遍歷第二個字串 ransomNote
        for char in ransomNote: # time: O(m)
            # 同樣判斷是否為英文字母
            if char.isalnum():
                # 統一轉為小寫
                char = char.lower()
                # 核心檢查：若字母不在字典中(負數)，或計數已歸零，代表庫存不足
                # 可以不需要小於，但是加上後如果其他地方出錯可以提升魯棒性(Robustness)
                if counts.get(char, 0) <= 0:
                    return False
                
                # 字母匹配成功，將計數減 1
                counts[char] -= 1

        return True
````

````python
magazine = "Our 2 bears live in a den 嗎"
ransomNote = "In our den 嗎"
Solution().canConstruct(ransomNote, magazine)
````

**時間複雜度: $O(n + m)$**  
**空間複雜度: $O(2n + m)$**

- 可讀性較高，且有機會提早退出

````python
class Solution:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        # 1. 資料標準化與清洗
        # 使用 List Comprehension 同步完成過濾與轉小寫，效率高
        magazine_clean = [char.lower() for char in magazine if char.isalnum()] # time, space: O(n)
        ransomNote_clean = [char.lower() for char in ransomNote if char.isalnum()] # time, space: O(m)
        
        counts = {} # space: O(n)

        # 2. 早期退出 (Early Return)
        # 如果 ransomNote 比 magazine 長，不可能組出來
        if len(magazine_clean) < len(ransomNote_clean):
            return False
        
        # 3. 建立頻率表 (Building Phase)
        for char in magazine_clean: # time: O(n)
            counts[char] = counts.get(char, 0) + 1

        # 4. 驗證與消耗階段 (Verification Phase)
        for char in ransomNote_clean: # time: O(m)
            # 核心檢查：若字母不在字典中(負數)，或計數已歸零，代表庫存不足
            # 可以不需要小於，但是加上後如果其他地方出錯可以提升魯棒性(Robustness)
            if counts.get(char, 0) <= 0:
                return False
            
            counts[char] -= 1

        return True
````

````python
magazine = "Our 2 bears live in a den 嗎"
ransomNote = "In our den 嗎"
Solution().canConstruct(ransomNote, magazine)
````

#### List

**時間複雜度: $O(n + m)$**  
**空間複雜度: $O(n + m)$**

- 只有在確定只會出現英文的情況下才能使用，有數字和中文的話我們沒辦法確定索引。只適用於 Leetcode 383. Ransom Note
- 空間更省：字典（dict）需要額外儲存 Hash Key 和處理碰撞（Collision）的空間；而陣列（list）只佔用連續的 26 個整數空間。
- 速度更快：ord(char) - ord("a") 的數學運算通常比字典的雜湊函數運算更快。

````python
class Solution:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        # 1. 資料清洗：只保留字母並統一轉為小寫
        magazine_clean = [char.lower() for char in magazine if char.isalpha()] # time, space: O(n)
        ransomNote_clean = [char.lower() for char in ransomNote if char.isalpha()] # time, space: O(m)

        # 2. 早期退出：如果清洗後的ransomNote 比 magazine 長，必不為 Anagram
        if len(magazine_clean) < len(ransomNote_clean):
            return False

        # 3. 初始化長度 26 的桶子 (Bucket)
        # 索引 0 代表 'a', 1 代表 'b' ... 25 代表 'z'
        counts = [0] * 26 # space: O(1)

        # 4. 統計 magazine_clean 中各字母出現次數
        for char in magazine_clean: # time: O(n)
            # 透過 ASCII 碼相減計算出對應的陣列索引
            counts[ord(char) - ord("a")] += 1

        # 5. 消耗並驗證 ransomNote_clean
        for char in ransomNote_clean: # time: O(m)
            char_index = ord(char) - ord("a")
            # 關鍵邏輯：如果在扣除前計數已為 0，代表 ransomNote 的字母過多或 magazine 沒出現過
            if counts[char_index] <= 0:
                return False
            
            counts[char_index] -= 1

        return True
````

````python
magazine = "Our bears live in a den"
ransomNote = "In our den"
Solution().canConstruct(ransomNote, magazine)
````
