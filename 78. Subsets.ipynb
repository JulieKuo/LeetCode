### 78. Subsets

```python
class Solution:
    def subsets(self, nums):
        # 將輸入數組排序，確保相同元素相鄰，方便後續處理
        self.nums = sorted(nums)
        # 記錄數組的長度
        self.n = len(nums)
        # 用於存儲最終結果的列表
        self.res = []
        # 執行深度優先搜索，起始點為0，初始路徑為空列表
        self.dfs(index = 0, path=[])
        
        return self.res
    
    def dfs(self, index, path):
        print("-" * 150)
        print(f"index = {index}, path = {path}")

        # 添加當前路徑到結果列表
        self.res.append(path)
        print(f"res = {self.res}")

        # 從當前索引開始遍歷數組元素
        for i in range(index, self.n):
            print(f"index = {index}, i+1 = {i+1} -> dfs")

            # 如果當前元素與前一個元素相同，則跳過該元素，避免生成重複的子集
            if (i > index) and (self.nums[i] == self.nums[i-1]):
                continue

            # 遞歸調用深度優先搜索，更新索引和路徑
            self.dfs(index = (i + 1), path = (path + [self.nums[i]]))
```

```python
nums = [1, 2, 2, 3]
Solution().subsets(nums)
```
