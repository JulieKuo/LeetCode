### 78. Subsets

**時間複雜度: $O(n \times 2^n)$**  
**空間複雜度: $O(n \times 2^n)$**

```python
from typing import List

class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        # 使用回溯法生成所有子集
        def backtrack(start, path): # time: O(2^n) # space: O(n)
            # 將當前路徑加入結果中
            # 使用path[:]創建一個新的列表,避免後續修改影響結果
            result.append(path[:])
            print(f"{result=}")

            # 從start開始遍歷,確保不會重複使用之前的元素
            for i in range(start, len(nums)): # time: O(n)
                print("-" * 100)
                # 將當前數字加入路徑
                path.append(nums[i])
                print(f"add: ({start=}, {i=}, {nums[i]=})-> {path=}")
                # 遞迴處理下一個位置,start設為i+1避免重複
                backtrack(start=i+1, path = path)
                # 回溯:移除最後加入的數字,嘗試其他可能
                path.pop()
                print(f"pop: ({start=}, {i=}, {nums[i]=}) -> {path=}")

            print("_" * 100)

        # 儲存所有子集的結果列表
        result = [] # space: O(n * 2^n)，共有2^n個子集，每個子集最多有n個元素
        # 從位置0開始回溯,初始路徑為空列表
        backtrack(start=0, path=[])

        return result
```

```python
nums = [] # [] # 2^0 = 1
nums = [1] # [[], [1]] # 2^1 = 2
nums = [1,2] # [[], [1], [1,2], [2]] # 2^2 = 4
```

```python
nums = [1,2,3]
Solution().subsets(nums)
```
