### 78. Subsets

- 時間複雜度: $O(n \times 2^n)$
  - 一個長度為 $n$ 的集合，其子集的總數是 $2^n$ 個（每個元素都有「選」與「不選」兩種可能）。在遞迴過程中，我們總共會進入 backtrack 函式 $2^n$ 次。
  - 每一次進入時，我們需要執行 list(path) 將當前的子集複製到結果中，這個複製操作在最壞情況下需要 $O(n)$ 的時間。
  - 因此，總時間複雜度為 $O(n \times 2^n)$。
- 空間複雜度: $O(n \times 2^n)$
    - 路徑 result:
      - 對於長度為 $n$ 的陣列，總共會產生 $2^n$ 個子集。
      - 簡單來說：每個數字在所有子集中，都有「選」與「不選」兩種命運，所以每個數字都會剛好出現在一半（$2^{n-1}$ 個）的子集裡。
      - 因為總共有 $n$ 個數字，每個人都出現 $2^{n-1}$ 次，所以所有子集的元素總長度加總為 $n \times 2^{n-1}$（例如 $n=3$ 時，總長度為 $3 \times 4 = 12$）。
    - 遞迴棧空間：
      - 最深為 $n$ 層，佔用 $O(n)$。
    - 忽略常數項後，最終所需的總空間為 $O(n \times 2^n)$。

```python
from typing import List

class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        # 使用回溯法生成所有子集
        def backtrack(start, path): # time: O(2^n) # space: O(n)
            # 將當前路徑加入結果中
            # 使用path[:]創建一個新的列表,避免後續修改影響結果
            result.append(path[:])
            print(f"{result=}")

            # 從start開始遍歷,確保不會重複使用之前的元素
            for i in range(start, len(nums)): # time: O(n)
                print("-" * 100)
                # 將當前數字加入路徑
                path.append(nums[i])
                print(f"add: ({start=}, {i=}, {nums[i]=})-> {path=}")
                # 遞迴處理下一個位置,start設為i+1避免重複
                backtrack(start=i+1, path = path)
                # 回溯:移除最後加入的數字,嘗試其他可能
                path.pop()
                print(f"pop: ({start=}, {i=}, {nums[i]=}) -> {path=}")

            print("_" * 100)

        # 儲存所有子集的結果列表
        result = [] # space: O(n * 2^n)，共有2^n個子集，每個子集最多有n個元素
        # 從位置0開始回溯,初始路徑為空列表
        backtrack(start=0, path=[])

        return result
```

```python
nums = [] # [] # 2^0 = 1
nums = [1] # [[], [1]] # 2^1 = 2
nums = [1,2] # [[], [1], [1,2], [2]] # 2^2 = 4
```

```python
nums = [1,2,3]
Solution().subsets(nums)
```
