### 55. Jump Game

* 時間複雜度: O(n)
* 空間複雜度: O(1)

```python
from typing import List

class Solution:
    def canJump(self, nums: List[int]) -> bool:
        # 紀錄目前能跳到的最遠索引位置，初始為 0
        max_reach = 0
        # 定義目標位置：陣列的最後一個索引（一開始就在索引 0，長度減 1）
        target = len(nums) - 1

        # 逐一檢查陣列中的每個位置與其跳躍力
        for idx, num in enumerate(nums):
            # 如果目前索引已經超過了我們能到達的最遠距離，代表無法再前進
            if idx > max_reach:
                return False

            # 更新最遠可達位置：取「目前的 max_reach」與「當前位置 + 可跳躍距離」的較大值
            max_reach = max(max_reach, idx + num)

            # 如果最遠可達位置已經碰到或超過目標索引，代表任務成功
            if max_reach >= target:
                return True

        # 如果跑完所有迴圈都能到達每個位置，則回傳 True
        return True
```

```python
nums = [2,3,1,1,4]
Solution().canJump(nums)
```
