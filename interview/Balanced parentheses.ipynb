```python
class Solution:
    def is_string_balanced(string):
        count = 0
        for i in string:
            if i == "(":
                count += 1
            elif i == ")":
                count -= 1
            # This is critical, if the student misses this one
            # the whole thing is wrong.
            if count < 0:
                return False
        return count == 0
```

```python
class Solution:
    def is_string_balanced_new(string):
        stack = []
        for i in string:
            stack.append(i)
            if i == ")":
                element = ""
                while element != "(":
                    # This is critical, if we miss this one the
                    # whole thing is wrong.
                    if len(stack) == 0:
                        return False
                    element = stack.pop()
        return "(" not in stack
```
