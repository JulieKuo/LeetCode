### Trie Prefix Count 

Implement a data structure to count the number of strings with a given prefix. Assume a large number of prefix counts will be requested for the same dictionary, and split your code to setup (performed once per dictionary) and count (performed per prefix).

- 時間複雜度
  - Setup (建立索引): $O(N \times L)$  
    - 每個單字的每個字元只會被走過一次。
  - Count (查詢前綴): $O(M)$
    - $M$ 是 prefix 的長度。我們不再需要像原本遞迴解法那樣去 countAll() 遍歷子樹。
- 空間複雜度 
  - 空間佔用: $O(T)$  
    - $T$ 是 Trie 中所有節點的總數。每個節點只存一個字典和一個整數。

```python
class TrieNode:
    def __init__(self):
        # 只需要紀錄子節點與經過次數
        self.children = {}
        self.pass_count = 0

class PrefixTree:
    def __init__(self, words: list[str]):
        """
        Setup: 執行一次，建立索引
        時間複雜度: O(N * L)，N 為單字數，L 為單字平均長度
        """
        self.root = TrieNode()
        for word in words:
            current = self.root
            # 根節點 pass_count 代表字典總單字量
            current.pass_count += 1
            for char in word:
                if char not in current.children:
                    current.children[char] = TrieNode()
                current = current.children[char]
                # 累加經過此節點的計數
                current.pass_count += 1

    def Count(self, prefix: str) -> int:
        """
        Count: 每次查詢前綴的數量
        時間複雜度: O(M)，M 為 prefix 長度
        """
        current = self.root
        for char in prefix:
            if char not in current.children:
                # 一旦斷掉，代表沒有任何單詞以此為前綴
                return 0
            current = current.children[char]
        
        # 走到 prefix 的結尾，直接回傳預存的計數
        return current.pass_count
```

```python
data_structure = PrefixTree(["app", "apple", "appetizer", "elephant"])
data_structure.Count("app") == 3
```
