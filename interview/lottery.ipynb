抽抽樂總共有五種獎項, 1,2 獎各只有一個, 3 獎有 2 個，4 獎有 3 個，5 獎有 7 個，請寫出一個程式可以「隨機」的取得「不重複」的禮物，且：  
1 獎中獎機率為 3%  
2 獎中獎機率為 7%  
3 獎中獎機率為 30%  
4 獎中獎機率為 20%  
5 獎中獎機率為 40%  
  
**這一題要麻煩大家不要生成隨機陣列！！！**  
重點請放在抽獎這件事情上  
抽獎的本質就是不重複的隨機抽取  
  
  
**不重複：是指禮物被抽走之後，就不能再被抽了！！！**  
例如一獎是汽車一台，被抽走了，就沒辦法再被抽了，因為獎品剛剛被抽走了，沒有了！！對吧？  
  
如果 4 獎是 3 隻 iPhone，iPhone A 被抽走了，給人了，理所當然就不能再被抽，但是，抽獎箱內還有 iPhone B 跟 iPhone C 可以抽！！  
  
白話的說：題目說的不重複隨機抽取，並不是說不可以連續抽二隻 iPhone，只是提醒大家注意被抽走的獎品本來就不應該再次被抽 ~  
  
**解題的方法很多種，但是，我們題目已經註明不要使用的方法就請大家不要使用~ 謝謝！ No means no！！！**   
  
**所以，顯示結果的範例：**  
（這裡只是示意，目的是避免大家花時間在生成隨機陣列）  
抽到 3 號獎 目前尚未抽取的獎勵為 [ 1 , 2 , 4 , 4 , 4 , 5 , 5 , 5 , 5]  
抽到 4 號獎 目前尚未抽取的獎勵為 [ 1 , 2 , 4 , 4 , 5 , 5 , 5 , 5]  
抽到 5 號獎 目前尚未抽取的獎勵為 [ 1 , 2 , 4 , 4 , 5 , 5 , 5]  
.....  
抽到禮物箱內沒有禮物為止  

```python
import random

prizes = [
    [1, 1, 0.03],
    [2, 1, 0.07],
    [3, 2, 0.30],
    [4, 3, 0.20],
    [5, 7, 0.40]
    ]

while len(prizes) > 0:
    print(f"prizes = {prizes}\n")

    print("="*100)
    print("<< 更新累積機率 >>")
    cumsum = [prizes[0][2]]
    print(f"cumsum = {cumsum}\n")
    
    for i in range(1, len(prizes)):
        print(f"(cumsum[i-1] = cumsum[{i}-1] = cumsum[{i-1}] = {cumsum[i-1]}), (prizes[i][2] = prizes[{i}][2] = {prizes[i][2]})")
        cumsum.append(cumsum[i-1] + prizes[i][2])
        print(f"cumsum = {cumsum}\n")

    while True:
        print("-"*100)
        print("<< 抽獎 >>")
        rand = random.uniform(0, cumsum[-1])
        print(f"rand (0 ~ {cumsum[-1]}) = {rand}\n")

        # 二分法降低時間複雜度(log n)
        left = 0
        right = len(cumsum) - 1
        while left < right:
            mid = (left + right) // 2
            print(f"origin: left = {left}, mid = {mid}, right = {right}")

            if cumsum [mid] < rand:
                print(f"(cumsum[mid] = {cumsum[mid]}) < (rand = {rand})")
                left = mid + 1
            else:
                print(f"(cumsum[mid] = {cumsum[mid]}) >= (rand = {rand})")
                right = mid
            print(f"new:    left = {left}, mid = {mid}, right = {right}\n")

        prize = prizes[right][0]
        print(f"抽中 {prize} 獎 !\n")

        print("<< 更新箱內獎項 >>")
        print(f"origin: prizes = {prizes}")
        prizes[right][1] -= 1
        print(f"new:    prizes = {prizes}\n")


        if prizes[right][1] == 0:
            print(f"<< {prize}獎已抽完，刪除箱內的{prize}獎 >>")
            prizes.pop(right)

            print(f"new: prizes = {prizes}")

            break
```
