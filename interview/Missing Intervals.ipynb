### Missing Intervals

#### Question

You are given a list of distinct integers from 0 to a value MAX_VALUE.

Write a function to produce a string that describes numbers missing from the input List of values. The items in the result should be sorted in ascending order and separated by commas. When a gap spans only one number, the item is the number itself; when a gap is longer, the item comprises the start and the end of the gap, joined with a minus sign.

#### Follow-Up Questions

Candidates may ask some or all of the following questions to clarify:
- Are the values in the input sorted?
  - No, the input values are not guaranted to be sorted.
- Is MAX_VALUE strictly positive?
  - Yes. MAX_VALUE will always be >0
- Are any of the input values <0 or >MAX_VALUE?
  - No. All input values in the list will be strictly integers from 0 <= x <= MAX_VALUE
- Are any of the input values repeated?
  - No. All input values are distinct (non-repeating).
- What if the input values list is empty?
  - Then all numbers between 0 and MAX_VALUE should be reported as missing.
- Can my code change the input values list?
  - Yes, you can change the input values list. It is preferrable to not change the input values list, but you can do so if needed.

#### Example

If the students asks for an example, offer this:

```py
MAX_VALUE = 1000
VALUES = [0, 1, 2, 50, 52, 75]

RESULT = "3-49,51,53-74,76-1000"
```

> [!NOTE]: The problem does not state that the input is necessarily sorted, and the student should not assume that without at least asking. (see follow-up questions above) The input we use in our simple example is sorted, to leave the question intentionally ambiguous.

#### Additional Example

Do not offer this example to students unless they ask for it.

In this example, the values are unsorted, and the result includes a range starting at 0, as well as a single value at MAX_VALUE (100).

```py
MAX_VALUE = 100
VALUE = [37, 36, 2, 99]

RESULT = "0-1,3-35,38-98,100"
```

- 時間複雜度：$O(N \log N)$
  - 推導：
    - `values.sort()`：採用 Timsort 演算法，耗時 $O(N \log N)$。
    - `values + [max_value + 1]`：建立新列表，耗時 $O(N)$。
    - `for` 迴圈遍歷：執行 $N+1$ 次，耗時 $O(N)$。
    - `",".join(missing_parts)`：合併字串，耗時 $O(N)$。
    - 總結：最高階項為 $O(N \log N)$。
- 空間複雜度：$O(N)$
  - 推導：
    - 排序輔助空間：Python 的 `sort()` 在最差情況下需要 $O(N)$ 的額外空間。
    - 列表串接：`values + [max_value + 1]` 創建了一個全新的列表，消耗 $O(N)$ 空間。
    - 結果儲存：`missing_parts` 列表在最極端情況下（例如所有數字都缺失且不連續）會儲存約 $O(N)$ 個字串片段。
    - 總結：整體輔助空間為 $O(N)$。

````python
class Solution:
    def find_uncovered_intervals(self, values: list[int], max_value: int) -> str:
        # 1. 處理邊界與排序
        # Time: O(N log N), Space: O(N)
        values.sort()
        
        missing_parts = []
        # previous 初始為 -1，用來偵測從 0 開始的缺口
        previous = -1 # Space: O(1)
        
        # 2. 核心邏輯：遍歷排序後的數組，並在最後巡檢到 max_value + 1
        # 這樣可以一次性處理完中間缺口與最後一個數字到 max_value 的缺口
        # Time: O(N) - 遍歷一次陣列
        # Space: O(N) - 創建了一個新的暫時列表 [max_value + 1] 以及儲存結果的 missing_parts
        for current in values + [max_value + 1]:
            # 如果兩個數字之間存在至少一個整數
            if current - previous > 1:
                start = previous + 1
                end = current - 1
                
                # 根據缺失範圍的長度決定格式
                if start == end:
                    missing_parts.append(str(start))
                else:
                    missing_parts.append(f"{start}-{end}")
            
            # 移動指針
            previous = current
            
        # 3. 合併結果
        # Time: O(N), Space: O(N)
        return ",".join(missing_parts)
````

````python
MAX_VALUE = 100
VALUE = [37, 36, 2, 99]

RESULT = "0-1,3-35,38-98,100"

Solution().find_uncovered_intervals(VALUE, MAX_VALUE)
````
