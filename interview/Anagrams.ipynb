#### Question

#### Follow-Up Questions
Candidates may ask some or all of the following questions to clarify:
- What is an anagram?
  - Two words are considered anagrams if the letters of one word can be rearranged to form the other.
- Do you have to use all the letters?
  - Yes. All the letters of one word must be used when forming the other word, with no missing, nor leftover letters.
- What characters are allowed?
  - All characters are allowed, whitespace and punctuation characters are ignored for the anagram computation.
- Does case matter to determine if strings are anagrams?
  - No.
