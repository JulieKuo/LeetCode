### 300. Longest Increasing Subsequence

**時間複雜度: O($n log{_2} n$)**  
**空間複雜度: O($n$)**



```python
from typing import List

class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        # 如果 nums 是空的，返回 0
        if not nums:
            return 0
        
        # tails[i] 表示長度為 i+1 的遞增子序列的最小結尾數字
        tails = [] # space: O(n)
        
        for num in nums: # time: O(n)
            # 使用二分搜索找到 num 應該插入的位置
            left, right = 0, len(tails)
            
            while left < right: # time: O(log n)
                mid = (left + right) // 2
                print(f"{left = }, {right = }, {mid = }")
                if tails[mid] < num:
                    left = mid + 1
                    print(f"(tail[{mid}]={tails[mid]}) < ({num=})")
                else:
                    right = mid
                    print(f"(tail[{mid}]={tails[mid]}) >= ({num=})")
                print(f"{left = }, {right = }, {mid = }")
                print("-" * 50)
            
            # 如果找到的位置等於 tails 的長度，表示 num 比所有結尾都大
            if left == len(tails):
                tails.append(num)
                print(f"({num=}) 比所有結尾都大, 新增到 tails 中")
            else:
                # 否則更新該位置的值
                tails[left] = num
                print(f"({num=}) 更新到 tails[{left}] 中")

            print(f"當前數字: {num}, tails: {tails}")
            print("_" * 50)
        
        return len(tails)
```

```python
nums = [10,9,2,5,3,7,101,18]
Solution().lengthOfLIS(nums)
```

**時間複雜度: O($n^2$)**  
**空間複雜度: O($n$)**

```python
from typing import List

class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        # 如果 nums 是空的，返回 0
        if not nums:
            return 0

        # 獲取 nums 的長度
        length = len(nums)
        
        # 表示每個數字自身就是長度為 1 的遞增子序列
        dp = [1] * length # space: O(n)
        print(f"{nums = }\n{dp   = }")

        for right in range(1, length): # time: O(n^2)
            print("_" * 50)
            for left in range(right):
                print("-" * 50)
                print(f"{left = }, {right = }")

                # 如果左邊的數字小於右邊的數字
                if nums[left] < nums[right]:
                    print(f"O: ({nums[left] = }) < ({nums[right] = })")
                    print(f"-> {(dp[left] + 1) = }, {dp[right] = }")

                    # 更新 dp[right] 為 dp[left] + 1 和當前 dp[right] 之間的較大值
                    dp[right] = max((dp[left] + 1), dp[right])
                    print(f"--> {dp[right] = },  {dp = }")
                else:
                    print(f"X: ({nums[left] = }) >= ({nums[right] = })")

        return max(dp)
```

```python
nums = [10,9,2,5,3,7,101,18]
Solution().lengthOfLIS(nums)
```
