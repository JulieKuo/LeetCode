### 279. Perfect Squares

**時間複雜度: O($n \times \sqrt{n}$)**  
**空間複雜度: O($n$)**

```python
class Solution:
    def numSquares(self, n: int) -> int:
        # dp[i] 代表數字i最少可以由幾個完全平方數組成
        dp = [float("inf")] * (n + 1) # space: O(n)
        # 0可以由0個完全平方數組成
        dp[0] = 0

        # 計算1到n每個數字最少需要幾個完全平方數
        for i in range(1, n + 1): # time: O(n)
            # 測試每個小於等於i的完全平方數
            j = 1
            while j * j <= i: # time: O(sqrt(n))
                print(f"{i=}, {j=}, {(j * j)=}")
                print(f"{dp[i]=}, {(dp[i - j * j] + 1)=}")
                
                # dp[i]可以由dp[i-j*j]加上一個j*j組成
                # 取目前dp[i]和新計算出的值的最小值
                dp[i] = min(dp[i], dp[i - j * j]  + 1)
                
                print(f"{dp[i]=}")
                print(f"{dp=}\n")
                
                j += 1

        # 返回n最少需要的完全平方數數量
        return dp[i]
```

```python
n = 12
Solution().numSquares(n)
```
