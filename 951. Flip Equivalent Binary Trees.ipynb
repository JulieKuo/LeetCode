### 951. Flip Equivalent Binary Trees

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
root1 = TreeNode(
    val = 1, 
    left = TreeNode(
        val = 2, 
        left = TreeNode(
            val = 4, 
            left = None, 
            right = None
            ), 
        right = TreeNode(
            val = 5, 
            left = TreeNode(
                val = 7, 
                left = None, 
                right = None
                ), 
            right = TreeNode(
                val = 8, 
                left = None, 
                right = None
                )
            )
        ), 
    right = TreeNode(
        val = 3, 
        left = TreeNode(
            val = 6, 
            left = None, 
            right = None
            ), 
        right = None
        ))

root2 = TreeNode(
    val = 1, 
    left = TreeNode(
        val = 3, 
        left = None, 
        right = TreeNode(
            val = 6, 
            left = None, 
            right = None
            )
        ), 
    right = TreeNode(
        val = 2, 
        left = TreeNode(
            val = 4, 
            left = None, 
            right = None
            ), 
        right = TreeNode(
            val = 5, 
            left = TreeNode(
                val = 8, 
                left = None, 
                right = None
                ), 
            right = TreeNode(
                val = 7, 
                left = None, 
                right = None
                )
            )
        ))

# Input: root1 = [1,2,3,4,5,6,null,null,null,7,8], root2 = [1,3,2,null,6,4,5,null,null,null,null,8,7]
# Output: true
```

```python
class Solution:
    def flipEquiv(self, root1, root2):
        
        return self.dfs(root1, root2)
        

    def dfs(self, match, flipper):
        # Both are None.
        if (not match) and (not flipper):
            print(f"True: (not {match}) and (not {flipper})")
            return True
        
        # Only one is None, so they don't match.
        if (not match) or (not flipper):
            print(f"False: (not {match}) or (not {flipper})")
            return False        
        
        print(f"match: {match.val}, flipper: {flipper.val}")

        # They both exist, but values of root don't match.
        if match.val != flipper.val:
            print(f"False: {match.val} != {flipper.val}")
            return False
        
        # No flip.
        regular = self.dfs(match.left, flipper.left) and self.dfs(match.right, flipper.right)
        print(f"regular = {regular}")
        
        # Flip.
        flipped = self.dfs(match.left, flipper.right) and self.dfs(match.right, flipper.left)
        print(f"flipped = {flipped}")
        print(f"flipper({flipper.val}) - {(regular or flipped)}: (regular = {regular}) or (flipped = {flipped})")
        
        return (regular or flipped)
```

```python
Solution().flipEquiv(root1, root2)
```
