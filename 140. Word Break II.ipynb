### 140. Word Break II

- 時間複雜度：$O(n^3 + 2^n * n)$  
  - 遞迴與切割：
    - 子問題數量為 $n$（字串長度）。
    - 每個子問題內有一個 $O(n)$ 的迴圈與 $O(n)$ 的字串切片（target_code），基本結構開銷為 $O(n^3)$。
  - 組合生成：最糟情況，如果字典包含所有可能的子字串（例如 s = "aaaa", dict = ["a", "aa", "aaa", "aaaa"]），組合的數量會呈現指數級 $O(2^n)$ 增長，且每個組合理由$O(n)$個字組成。  
- 空間複雜度：$O(m + n + 2^n \cdot n)$  
  - 字典集合: $O(m)$，儲存集合單字所佔用的空間。
  - 遞迴堆疊：最大深度為 $n$，佔用 $O(n)$。  
  - 記憶化快取 (memo)：這是主要的空間消耗。我們必須在 memo 中儲存所有可能的組合字串。在最糟情況下，組合數為 $2^{n-1}$，且每個句子長度約為 $n$

```python
from typing import List

class Solution:
    def wordBreak(self, s: str, wordDict: List[str]) -> List[str]:
        # 將字典轉換為集合，提升查詢速度至 O(1)
        word_set = set(wordDict)  # space: O(m)。
        # 建立記憶化快取：Key 為起始索引，Value 為從該索引開始能組成的所有句子清單
        memo = {}

        # time: O(n^3), space: O(n)
        # time: 子問題數量：O(n) (共有 n 個 start)
        # time: 單個子問題內的運算：迴圈 O(n) * 切片 O(n) = O(n^2)
        # time: 不含結果拼接的總開銷：O(n * n^2) = O(n^3)
        def dfs(start): # time: O(n), space: O(n)
            # 若此起始索引已計算過，直接從快取回傳結果
            if start in memo:
                return memo[start]

            # 終止條件：若起始索引到達字串末尾，回傳包含空字串的清單作為拼接基礎
            if start == len(s):
                return [""]
            
            # 用於儲存從當前 start 開始能產生的所有組合
            result = []

            # 遍歷所有可能的切割點 (end)
            for end in range(start + 1, len(s) + 1): # time: O(n)
                # 切出當前的前綴單字
                word = s[start:end] # time: O(n)
                
                # 如果該單字存在於字典中
                if word in word_set:
                    # 遞迴取得剩餘字串的所有拆分組合
                    sub_sentences = dfs(end)
                    
                    # 將當前單字與後續的所有組合進行拼接
                    for sub in sub_sentences:
                        if sub == "":
                            # 若後續為空（代表已到字串末尾），直接加入單字
                            result.append(word)
                        else:
                            # 否則中間加上空格進行拼接
                            result.append(word + " " + sub)

            # 將計算完畢的所有組合存入記憶化快取中
            # time, space: O(2^n * n)，最差情況會有指數成長的組合，每個組合有 n 個字
            # sub_sentences 是空的，就不會有組合。不是每個 dfs 都包含 O(2^n) 操作，而是 O(n^3) 基礎開銷 加上 額外開銷 O(2^n)
            memo[start] = result
            # 回傳當前層級的所有拆分結果
            return result

        # 從索引 0 開始執行深度優先搜尋
        return dfs(0)
```

```python
s = "pineapplepenapple"
wordDict = ["apple","pen","applepen","pine","pineapple"]
# Output: ["pine apple pen apple","pineapple pen apple","pine applepen apple"]

Solution().wordBreak(s, wordDict)
```
