### 131. Palindrome Partitioning

```python
from collections import defaultdict

class Solution(object):
    def __init__(self):
        # 使用 defaultdict 創建一個空的字典，預設值為空串列
        self.memory = defaultdict(list)

    def partition(self, s):
        print("-" * 150)
        print(f"s = {s}")
        
        # 如果輸入字串為空，回傳包含一個空串列的串列
        if not s:
            print("return [[]]")
            return [[]]

        # 如果字串已經處理過，直接回傳記憶中的結果以節省運算時間
        if s in self.memory:            
            print(f"return (memory[{s}] = {self.memory[s]})")
            return self.memory[s]

        # 初始化存放結果的串列
        answer = []
        
        # 逐一檢查字串的不同前綴
        for i in range(1, len(s) + 1):
            prefix = s[:i]
            print(f"i = {i}, prefix = {prefix}")
            
            # 如果字串的前綴是回文串
            if prefix == prefix[::-1]:
                suffix0 = s[i:]
                print(f"-> prefix = {prefix}, suffix0 = {suffix0} -> dfs")
                
                # 遞迴處理剩餘的後綴部分
                suffixes = self.partition(suffix0)
                print(f"--> prefix = {prefix}, suffixes = {suffixes}")
                
                # 將當前的回文串前綴與後綴組合起來並加入結果串列
                for suffix in suffixes:
                    print(f"--> ([prefix] = [{prefix}]) + (suffix = {suffix})")
                    answer.append([prefix] + suffix)
            else:
                print("-> not")

        # 將當前字串及其對應的分割結果存入記憶中
        self.memory[s] = answer
        print(f"memory[{s}] = {self.memory[s]}")
        
        # 回傳最終的分割結果
        print(f"return (answer = {answer})")
        return answer
```

```python
s = "aabcb"
Solution().partition(s)
```
