## 1991. Find the Middle Index in Array

- index左邊的加總 = index右邊的加總
- 多個解時答案為最小 index

- 時間複雜度：$O(N)$
- 空間複雜度：$O(1)$

```python
from typing import List

class Solution:
    def findMiddleIndex(self, nums: List[int]) -> int:
        # 初始化左側總和為 0
        left_sum = 0 # space: O(1)
        # 初始時，右側總和設為整個陣列的加總
        right_sum = sum(nums) # space: O(1)

        # 使用 enumerate 同時取得索引 (idx) 與數值 (num)
        for idx, num in enumerate(nums): # time: O(n)
            # 當前數字 num 被視為「中心點（Pivot）」，不屬於左也不屬於右
            # 因此先從右側總和中扣除當前數字
            right_sum -= num
            
            # 判斷此時左側累計和是否等於剩餘的右側和
            if left_sum == right_sum:
                # 若相等，則找到平衡索引，直接回傳（確保是最小索引）
                return idx
            
            # 若不相等，將當前數字加入左側累計和，繼續檢查下一個索引
            left_sum += num

        # 若遍歷完整個陣列都沒有符合條件的索引，回傳 -1
        return -1
```

```python
nums = nums = [1,7,3,6,0,0,5,6]
Solution().findMiddleIndex(nums)
```
