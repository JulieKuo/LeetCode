## 1991. Find the Middle Index in Array

- index左邊的加總 = index右邊的加總
- 多個解時答案為最小 index

- 時間複雜度：$O(N)$
- 空間複雜度：$O(1)$

```python
from typing import List

class Solution:
    def findMiddleIndex(self, nums: List[int]) -> int:
        # 初始化左側總和為 0
        left_sum = 0
        
        # 先計算出整個陣列的總數字和
        total_sum = sum(nums)

        # 使用 enumerate 同時取得索引 (idx) 與數值 (num)
        for idx, num in enumerate(nums):
            # 右側總和 = 總數字和 - 左側累計和 - 當前中心點數字
            right_sum = total_sum - left_sum - num
            
            # 檢查左側總和是否等於右側總和
            if left_sum == right_sum:
                # 因為是從左往右遍歷，第一個符合的一定是最左邊的 middleIndex
                return idx
            
            # 若不相等，將當前數字累加到左側總和，繼續檢查下一個位置
            left_sum += num

        # 若遍歷完所有索引都找不到平衡點，依題意回傳 -1
        return -1
```

```python
nums = nums = [1,7,3,6,0,0,5,6]
Solution().findMiddleIndex(nums)
```
