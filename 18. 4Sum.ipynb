### 18. 4Sum

```python
class Solution:
    def fourSum(self, nums, target):
        def findNsum(left, right, target, n, result, results):
            print(f"{'-'*1000}")
            print(f"n = {n}, target = {target}, nums = {nums}, result = {result}, results = {results}\n")
            print(f"original({n}): left = {left}, right = {right}")
            if (right-left+1 < n) or (n < 2) or (target < nums[left]*n) or (target > nums[right]*n):
                # (right-left+1 < n) 剩下的元素是否比要找的元素還少
                # (n < 2) 不需要找
                # (target < nums[left]*n) n個元素加總要等於target，如果最小的值*n已經比target大，加總不可能再小了
                # (target > nums[right]*n) n個元素加總要等於target，如果最大的值*n已經比target小，加總不可能再大了
                print("<< early termination >>")
                if (left < len(nums)) and (right < len(nums)):
                    print(f"((right-left+1 = {right-left+1}) < (n = {n})) or ((n = {n}) < 2) or ((target = {target}) < (nums[left]*n = {nums[left]}*{n})) or ((target = {target}) > (nums[right]*n = {nums[right]}*{n}))")
                elif (left < len(nums)):
                    print(f"((right-left+1 = {right-left+1}) < (n = {n})) or ((n = {n}) < 2) or ((target = {target}) < (nums[left]*n = {nums[left]}*{n}))")
                elif (right < len(nums)):
                    print(f"((right-left+1 = {right-left+1}) < (n = {n})) or ((n = {n}) < 2) or ((target = {target}) > (nums[right]*n = {nums[right]}*{n}))")
                else:
                    print(f"((right-left+1 = {right-left+1}) < (n = {n})) or ((n = {n}) < 2)")
                
                return
            
            if n == 2:
                while (left < right):
                    total = nums[left] + nums[right]
                    print(f"\nupdate({n}): left = {left}, right = {right}")
                    print(f"nums[left] = {nums[left]}, nums[right] = {nums[right]}, s = {total}")
                    if (total == target):
                        results.append(result + [nums[left], nums[right]])
                        print(f"results = {results}")
                        print(f"(total = {total}) == (target = {target})")
                        while (left < right) and (nums[left] == nums[left+1]):  # 該元素與前一元素相同時，即跳過，避免出現重複的組合
                            print(f"((left = {left}) < (right = {right})) and ((nums[left] = {nums[left]}) == (nums[left-1] = {nums[left-1]}))")
                            left += 1
                            print(f"same({n}): left = {left}, right = {right}")
                        while (left < right) and (nums[right] == nums[right-1]):  # 該元素與前一元素相同時，即跳過，避免出現重複的組合
                            print(f"((left = {left}) < (right = {right})) and ((nums[right] = {nums[right]}) == (nums[right-1] = {nums[right-1]}))")
                            right -= 1
                            print(f"same({n}): left = {left}, right = {right}")
                        
                        left += 1
                        right -= 1
                        print(f"new({n}): left = {left}, right = {right}")
                    elif (total < target):
                        left += 1
                        print(f"(total = {total}) < (target = {target})")
                        print(f"new({n}): left = {left}, right = {right}")
                    else:
                        right -= 1
                        print(f"(total = {total}) > (target = {target})")
                        print(f"new({n}): left = {left}, right = {right}")
            else:
                for i in range(left, right+1):
                    if (i == left) or (nums[i-1] != nums[i]):  
                        # (i == left) 第一個元素前面沒有值可以比較，單設一個條件
                        # (nums[i-1] != nums[i]) 該元素與前一元素不相同時，才可進入findNsum查找組合，避免出現重複的組合
                        findNsum(i+1, right, target-nums[i], n-1, result+[nums[i]], results)


        nums.sort()
        results = []
        findNsum(0, len(nums)-1, target, 4, [], results)
        return results
```

```python
nums = [1, 1,0,-1,0,-2,2]
target = 0

Solution().fourSum(nums, target)
```
