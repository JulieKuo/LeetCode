### 526. Beautiful Arrangement

```python
class Solution:
    def countArrangement(self, n: int) -> int:
        self.res = 0 # 初始化結果
        perm = [i for i in range(1, n+1)] # 生成包含1到n的排列
        self.n = n

        self.dfs(i = 1, perm = perm) # 執行深度優先搜尋

        return self.res
    
    def dfs(self, i, perm):
        print("-" * 150)
        print(f"i = {i}, perm = {perm}")

        # 如果已經遍歷完整個排列，結果加一
        if i == (self.n + 1):
            self.res += 1
            print(f"result = {self.res}")

            return

        # 遍歷目前的排列
        for j in range(len(perm)):
            print(f"-> i = {i}, j = {j}, perm[j] = {perm[j]}")

            # 如果 (perm[j] 能整除 i) 或者 (i 能整除 perm[j])
            if (not (perm[j] % i)) or (not (i % perm[j])):
                print("divisible")
                # 遞迴呼叫下一層，i加一，同時刪除成功的元素
                self.dfs(i = (i + 1), perm = (perm[:j] + perm[j+1 :]))

            
            print("_" * 150)
```

```python
n = 3
# [1, 2, 3]
# [2, 1, 3]
# [3, 2, 1]
# output: 3
```

```python
Solution().countArrangement(n)
```
