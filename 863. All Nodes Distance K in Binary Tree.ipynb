### 863. All Nodes Distance K in Binary Tree

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
root = TreeNode(
    val = 3,
    left = TreeNode(
        val = 5,
        left = TreeNode(
            val = 6,
            left = None,
            right = None
        ),
        right = TreeNode(
            val = 2,
                left = TreeNode(
                val = 7,
                left = None,
                right = None
            ),
            right = TreeNode(
                val = 4,
                left = None,
                right = None
            )
        ),
    ),
    right = TreeNode(
        val = 1,
        left = TreeNode(
            val = 0,
            left = None,
            right = None
        ),
        right = TreeNode(
            val = 8,
            left = None,
            right = None
        )
    )
)
target = root.left
k = 2

# Input: root = [3,5,1,6,2,0,8,null,null,7,4], target = 5, k = 2
# Output: [7,4,1]
```

```python
class Solution:
    def __init__(self):
        self.parentMap = {} # 存儲每個節點的父節點
        self.visited = set() # 跟蹤已經訪問過的節點
        self.result = [] # 存儲結果，即距離目標節點特定距離的節點

    def distanceK(self, root, target, k):
        self.k = k

        self.getParent(root, None) # 呼叫getParent函數，遍歷樹並將每個節點的父節點存儲在字典中
        self.getNodes(target, 0) # 呼叫getNodes函數，查找距離目標節點特定距離的節點

        return self.result

    def getParent(self, node, parent):
        if not node:
            return

        self.parentMap[node] = parent  # 子節點為key，父節點為value

        self.getParent(node.left, node)  # 遞迴處理左子樹
        self.getParent(node.right, node)  # 遞迴處理右子樹

    def getNodes(self, node, distance):
        print("-" * 200)
        if (not node) or (node in self.visited) or (distance > self.k):
            print("(not node) or (node in self.visited) or (distance > self.k)")
            return

        print(f"node = {node.val}, distance = {distance}, k = {k}")
        self.visited.add(node)  # 將當前節點標記為已訪問
        if distance == self.k:
            self.result.append(node.val)  # 當距離為k時，將節點值添加到結果中            
            print(f"result = {self.result}")

        print(f"{node.val} - left")
        self.getNodes(node.left, distance + 1)  # 遞迴處理左子樹，距離+1
        print(f"{node.val} - right")
        self.getNodes(node.right, distance + 1)  # 遞迴處理右子樹，距離+1
        print(f"{node.val} - parent")
        self.getNodes(self.parentMap[node], distance + 1)  # 遞迴處理父節點，距離+1
```

```python
Solution().distanceK(root, target, k)
```
