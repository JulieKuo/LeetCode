### 190. Reverse Bits

- 時間複雜度：$O(1)$  
- 空間複雜度：$O(1)$  

```python
class Solution:
    def reverseBits(self, n: int) -> int:
        # 初始化結果為 0
        result = 0
        
        # 我們必須處理完所有的 32 個位元
        for i in range(32):
            # 第一步：先將結果向左移一位，騰出最右邊的空間
            # 就像是排隊時，前面的人往左跨一格，空出最後一個位置
            result = result << 1
            
            # 第二步：取出 n 目前的最右邊那一位 (Least Significant Bit)
            # 使用 & 1 操作，如果最後一位是 1 則得到 1，是 0 則得到 0
            current_bit = n & 1
            
            # 第三步：將取出的這一位放到 res 剛才騰出的空位上
            # 使用 | (OR) 運算或直接相加都可以
            result = result | current_bit
            
            # 第四步：將 n 向右移一位
            # 意思是「這一位我處理完了，丟掉它，讓下一位變成最後一位」
            n = n >> 1
            
        # 當 32 次迴圈跑完，res 就裝滿了反轉後的位元
        return result
```

```python
n = 43261596
# 964176192
Solution().reverseBits(n)
```
