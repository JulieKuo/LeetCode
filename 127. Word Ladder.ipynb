## 127. Word Ladder

```python
from collections import defaultdict
from collections import deque

class Solution:
    def ladderLength(self, beginWord, endWord, wordList):
        if (endWord not in wordList) or (not endWord) or (not beginWord) or (not wordList):
            return 0
            
        L = len(beginWord)
        all_combo_dict = defaultdict(list)
        for word in wordList:
            for i in range(L):
                all_combo_dict[word[:i] + "*" + word[i+1:]].append(word)
        print(f"all_combo_dict = {all_combo_dict}")

        queue = deque([(beginWord, 1)])
        print(f"queue = {queue}")
        visited = set()
        visited.add(beginWord)
        print(f"visited = {visited}")

        while queue:
            print(f"{'_'*200}")
            print(f"init: queue = {queue}")
            current_word, level = queue.popleft()     
            print(f"current_word = {current_word}, level = {level}")
            
            for i in range(L):
                print(f"{'-'*100}")
                intermediate_word = current_word[:i] + "*" + current_word[i+1:]
                print(f"intermediate_word = {intermediate_word}")
                print(f"all_combo_dict[intermediate_word] = {all_combo_dict[intermediate_word]}")
                for word in all_combo_dict[intermediate_word]:
                    if word == endWord:
                        print(f"(word = {word}) == (endWord = {endWord})")
                        print(f"level = {level + 1}")
                        return level + 1
                    
                    if word not in visited:
                        print(f"-> (word = {word}) not in (visited = {visited})")
                        print(f"init: visited = {visited}")
                        visited.add(word)
                        print(f"end:  visited = {visited}")
                        queue.append((word, level + 1))
                        print(f"end:  queue = {queue}")
                    else:                        
                        print(f"-> (word = {word}) in (visited = {visited})")
        return 0
```

```python
beginWord = "hit"
endWord = "cog"
wordList = ["hot","dot","dog","lot","log","cog"]
Solution().ladderLength(beginWord, endWord, wordList)
```
