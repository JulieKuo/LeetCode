### 698. Partition to K Equal Sum Subsets

nums = [4,3,2,3,5,2,1]; k = 4

```python
nums = [1, 1,1, 2,7]; k = 3  
# Output: false
```

```python
nums = [3522,181,521,515,304,123,2512,312,922,407,146,1932,4037,2646,3871,269]
k = 5
```

```python
class Solution:
    def canPartitionKSubsets(self, nums, k):
        total = sum(nums) # 計算陣列中所有數字的總和

        # 如果總和不能被k整除，無法分割成k個子集，直接返回False
        if (total % k):
            return False

        self.subset_sum = (total // k) # 計算每個子集應該擁有的總和

        # 初始化k個子集
        self.subsets = [0] * k

        # 降序以優化搜索過程
        self.nums = sorted(nums, reverse = True)
        print(f"nums = {self.nums}")
        
        self.k = k
        self.n = len(nums)
        
        return self.dfs(i = 0) # 執行深度優先搜索，從第一個數字開始

    def dfs(self, i):
        print("_" * 150)
        # 如果已經遍歷完所有數字，返回True表示成功分割
        if (i == self.n):
            print(f"True -> (i = {i}) == (len(nums) = {len(self.nums)})")
            return True

        # 嘗試將當前數字放入每個子集中
        for j in range(self.k):
            print(f" j = {j}, i = {i}, nums[i] = {self.nums[i]}")
            # 如果將當前數字加入第j個子集後，總和不超過所需總和
            if (self.subsets[j] + self.nums[i]) <= self.subset_sum:
                print(f"((subsets[j] = {self.subsets[j]}) + (nums[i] = {self.nums[i]})) <= (subset_sum = {self.subset_sum})")
                self.subsets[j] += self.nums[i] # 將當前數字加入第j個子集
                print(f"subsets = {self.subsets}")

                print(f"dfs -> i = {i + 1}")
                # 遞歸搜索下一個數字
                if self.dfs(i = (i + 1)):
                    print(f"True -> j = {j}, i = {i + 1}")
                    return True
                
                print("_" * 150)
                print(f"False -> j = {j}, i = {i}, nums[i] = {self.nums[i]}")

                # dfs失敗，將當前數字從第j個子集中移除，試著加入到第j個子集後的子集中
                self.subsets[j] -= self.nums[i]
                print(f"subsets = {self.subsets}")

            # 如果subsets[j]為0，表示遞迴失敗，subsets[j+1:]也會失敗，直接跳出
            if self.subsets[j] == 0:
                print(f"False -> self.subsets[{j}] == 0")
                break

            
            print("-" * 150)


        # 如果所有嘗試都失敗，返回False
        return False
```

```python
Solution().canPartitionKSubsets(nums, k)
```
