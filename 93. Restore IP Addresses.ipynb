### 93. Restore IP Addresses

```python
s = "25525511135"
# ["255.255.11.135","255.255.111.35"]

s = "0000"
# ["0.0.0.0"]

s = "101023"
# ["1.0.10.23","1.0.102.3","10.1.0.23","10.10.2.3","101.0.2.3"]
```

```python
class Solution:
    def restoreIpAddresses(self, s: str) -> list:
        ans = []
        # 從第一個區段開始構建IP位址
        self.solve(ans = ans, output = "", index = 0, s = s, dots = 0)
        
        return ans

    def solve(self, ans, output, index, s, dots):
        print("-" * 150)
        print(f"ans = {ans}")
        print(f"s = {s}, index = {index}, dots = {dots}, output = {output}")
        # 當點的數量為3時，檢查最後一個區段是否為有效的IP位址
        if dots == 3:
            print(f"({dots}) - {self.valid(s[index:])}: s[index:] = s[{index}:] = {s[index:]}")
            if self.valid(s[index:]):
                ans.append(output + s[index:])
                print(f"-> ans = {ans}")
            print(f"-> dots({dots-1}) refind")
            return
        
        # 遍歷IP位址中每個區段的可能性
        for i in range(index, min(index+3, len(s))):
            print(f"({dots}) - {self.valid(s[index:(i+1)])}: s[index:(i+1)] = s[{index}:({i}+1)] = s[{index}:{i+1}] = {s[index:i+1]}")
            if self.valid(s[index:(i+1)]):
                new_output = output + s[index:i+1] + '.'
                print(f"new_output = {new_output}")
                self.solve(ans = ans, output = new_output, index = i+1, s = s, dots = dots+1) # 遞迴處理下一個區段
    
    def valid(self, temp: str) -> bool:
        # 如果長度為0，則不是有效的IP位址
        if (len(temp) == 0):
            return False
        
        # 如果長度大於1且第一個數字為0，則不是有效的IP位址
        if (len(temp) > 1) and (temp[0] == '0'):
            return False
        
        # 如果數字大於255，則不是有效的IP位址
        if (len(temp) != 0) and (int(temp) > 255):
            return False
        
        # 是有效的IP位址
        return True
```

```python
Solution().restoreIpAddresses(s)
```
