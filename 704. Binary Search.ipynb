## 704. Binary Search

<img src="img/704.jpg" style="zoom:50%"/>

- 時間複雜度： $O(log n)$
  - 每次迴圈我們都會將搜尋的陣列範圍砍半（$n \rightarrow \frac{n}{2} \rightarrow \frac{n}{4} \dots \rightarrow 1$）。在最壞的情況下，最多只需要比較 $\log_2 n$ 次，因此時間複雜度為 O(log n)。
- 空間複雜度： $O(1)$
  - 整個演算法只使用了常數個變數（left, right, mid）來記錄索引位置，並沒有配置任何與輸入陣列大小成正比額外記憶體空間。

```python
class Solution:
    def search(self, nums: list[int], target: int) -> int:
        # 初始化左右雙指標，定義搜尋範圍為閉區間 [left, right]
        left, right = 0, len(nums) - 1
        
        # 當左指標小於或等於右指標時，代表搜尋空間內仍有元素
        while left <= right:
            # 計算中間索引，使用此寫法可有效防止在其他語言中潛在的大數溢位問題
            mid = left + (right - left) // 2
            
            # 找到目標值，直接回傳其索引
            if nums[mid] == target:
                return mid
            # 中間值小於目標值，說明目標在右半邊，將左邊界往右移
            elif nums[mid] < target:
                left = mid + 1
            # 中間值大於目標值，說明目標在左半邊，將右邊界往左移
            else:
                right = mid - 1
                
        # 若跳出迴圈仍未找到目標值，依題意回傳 -1
        return -1
```

```python
nums =[-1,0,3,5,9,12]
target = 9

Solution().search(nums, target)
```
