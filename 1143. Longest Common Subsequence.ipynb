### 1143. Longest Common Subsequence

每次只需要用到當前行和上一行的數據，所以可以只使用一行和一個變數來儲存數據，更優化空間複雜度

**時間複雜度: $O(m \times n)$**  
**空間複雜度: $O(min(m,n))$**

```python
class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        # 優化：確保 text2 始終是較短的字串，這樣 dp 陣列長度最小，節省空間
        if len(text1) < len(text2):
            text1, text2 = text2, text1

        m = len(text1)
        n = len(text2)

        # 建立一維 DP 陣列，長度為較短字串長度 + 1
        dp = [0] * (n + 1)

        # 外層迴圈遍歷 text1 的字元 (row)
        for row in range(1, m + 1):
            # top_left_corner 代表二維表格中的 dp[row-1][col-1] (對角線左上方)
            top_left_corner = 0 
            
            # 內層迴圈遍歷 text2 的字元 (col)
            for col in range(1, n + 1):
                # 在更新 dp[col] 之前，它目前存的是「正上方」的值 (即上一列的同一個位置)
                # 我們將其暫存起來，因為對下一個 col 來說，這個值就是它的「左上方」
                temp = dp[col]

                # 如果兩個字元匹配
                if text1[row - 1] == text2[col - 1]:
                    # 新的 LCS 長度 = 左上方值 + 1
                    dp[col] = top_left_corner + 1
                else:
                    # 如果不匹配，取「上方」(目前的 dp[col]) 或「左方」(剛更新完的 dp[col-1]) 的最大值
                    dp[col] = max(dp[col], dp[col - 1])
                
                # 更新 top_left_corner，讓下一個 col 迴圈可以使用目前的 temp 作為左上方值
                top_left_corner = temp

        # 回傳一維陣列的最後一個元素
        return dp[n]
```

```python
text1 = "abcde"
text2 = "ace"
print(Solution().longestCommonSubsequence(text1, text2))
```

每次只需要用到當前行和上一行的數據，所以可以只使用兩行來儲存數據，將空間複雜度從 O(m×n) 優化到 O(min(m,n))

**時間複雜度: $O(m \times n)$**  
**空間複雜度: $O(min(m,n) \times 2)$**

```python
class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        # 確保 text2 是較短的字符串，以減少空間使用
        if len(text1) < len(text2):
            text1, text2 = text2, text1
            
        m, n = len(text1), len(text2)
        
        # 只使用兩行的DP表格
        dp = [[0] * (n + 1) for _ in range(2)]  # space: O(2 * min(m,n)) = O(min(m,n))
        
        # 填充DP表格
        for i in range(1, m + 1):  # time: O(m*n)
            for j in range(1, n + 1):
                if text1[i-1] == text2[j-1]:
                    dp[i % 2][j] = dp[(i-1) % 2][j-1] + 1
                else:
                    dp[i % 2][j] = max(dp[(i-1) % 2][j], dp[i % 2][j-1])
                
                
                print(f"{i=}, {j=}")
                print(f"{text1=}, {text2=}")
                print(f"text1[i-1]=text1[{i-1}]={text1[i-1]}, text2[j-1]=text2[{j-1}]={text2[j-1]}")
                print(f"當前行:{(i%2)=}, 上一行:{((i-1)%2)=}")
                for row in dp: print(row)
                print("-"*100)
        
        return dp[m % 2][n]
```

```python
text1 = "abcde"
text2 = "ace"
print(Solution().longestCommonSubsequence(text1, text2))
```

**時間複雜度: O($m \times n$)**  
**空間複雜度: O($m \times n$)**

```python
class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        m, n = len(text1), len(text2)
        # 創建 DP 表格，初始化為 0
        dp = [[0] * (n + 1) for _ in range(m + 1)] # space: O(m*n)
        
        # 填充 DP 表格
        for i in range(1, m + 1): # time: O(m*n)
            for j in range(1, n + 1):
                if text1[i-1] == text2[j-1]:
                    # 如果字符相同，則左上角值+1
                    dp[i][j] = dp[i-1][j-1] + 1
                else:
                    # 如果字符不同，則取上一行或左邊欄的兩者最大值
                    dp[i][j] = max(dp[i-1][j], dp[i][j-1])
                
                print(f"{i=}, {j=}")
                print(f"{text1=}, {text2=}")
                print(f"text1[i-1]=text1[{i-1}]={text1[i-1]}, text2[j-1]=text2[{j-1}]={text2[j-1]}")
                for row in dp: print(row)
                print("-"*100)
        
        return dp[m][n]
```

```python
text1 = "abcde"
text2 = "ace"
print(Solution().longestCommonSubsequence(text1, text2))
```
