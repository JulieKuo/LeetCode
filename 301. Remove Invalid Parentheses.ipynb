### 301. Remove Invalid Parentheses

```python
s = "()())()"
# Output: ["(())()","()()()"]

s = "(a)()))()"
# Output: ["(a())()","(a)()()"]
```

```python
class Solution:
    def removeInvalidParentheses(self, s: str) -> list:        
        stack = self.isValid(s)[1]  # 使用 isValid 函數檢查括號是否有效，獲取無效括號的資訊
        left = sum([1 for val in stack if val[1] == "("])  # 無效左括號的數量
        right = len(stack) - left  # 無效右括號的數量
        print(f"stack = {stack}, left = {left}, right = {right}")
        
        self.res, self.visited = [], set()
        self.dfs(s, left, right)  # 開始深度優先搜索
        return self.res

    def isValid(self, s):
        stack = [] # 儲存無效括號
        for i, ch in enumerate(s):
            if ch == '(':
                stack.append((i, ch)) # 左括號直接存入
            elif ch == ')':
                if stack and (stack[-1][1] == '('): # 如果前面有左括號，能與右括號成對，就把左括號刪掉
                    stack.pop()
                else: # 否則存入無效的右括號
                    stack.append((i, ch)) 
        return (len(stack) == 0), stack
    
    def dfs(self, s, left, right):
        self.visited.add(s) # 儲存已訪問過的字串
        print("-" * 150)        
        print(f"s = '{s}', left = {left}, right = {right}")
        print(f"visited = {self.visited}")

        if (left == 0) and (right == 0):
            print(f"--> (left == 0) and (right == 0)")
            if self.isValid(s)[0]: 
                self.res.append(s)  # 找到有效的括號組合，將其添加到結果中
                print(f"--> (self.isValid(s)[0] = {self.isValid(s)[0]})")
                print(f"--> res = {self.res}")
            print("_" * 150)
            return

        for i, ch in enumerate(s):
            print(f"s = '{s}', i = {i}, ch = {ch}, left = {left}, right = {right}")
            if ch not in ["(", ")"]:  # 如果是其他字符，則忽略
                continue

            if ((ch == '(') and (left == 0)) or ((ch == ')') and (right == 0)):  # 如果 left == 0，則沒有多餘的'('，刪除'('會導致不平衡，因此跳過，反之亦然。
                continue
            
            print(f"-> (s[:{i}] + s[{i+1}:] = '{s[:i] + s[i+1:]}') not in (visited = {self.visited}) = {(s[:i] + s[i+1:]) not in self.visited}")
            if (s[:i] + s[i+1:]) not in self.visited:
                self.dfs(s = (s[:i] + s[i+1:]), left = (left - (ch == '(')), right = (right - (ch == ')'))) # 如果ch為"("，刪掉"("及代表無效左括號少一個
```

```python
Solution().removeInvalidParentheses(s)
```
