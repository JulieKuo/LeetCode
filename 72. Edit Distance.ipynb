## 72. Edit Distance

```python
class Solution:
  def minDistance(self, word1: str, word2: str) -> int:
    print(f"{word1 = }; {word2 = }")

    # 取得兩個字串的長度
    m = len(word1)
    n = len(word2)
    
    # 建立動態規劃表格，row數為 m+1，column數為 n+1
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    print(f"{dp = }")

    # 初始化每個row的第0個column (word2為空時，word1變得跟word2一樣所需要的操作次數)
    for r in range(1, m + 1):
      dp[r][0] = r
      print(f"initialize col({word1[r-1]}): {dp = }")

    # 初始化每個column的第0個row (word1為空時，word2變得跟word1一樣所需要的操作次數)
    for c in range(1, n + 1):
      dp[0][c] = c
      print(f"initialize row({word2[c-1]}): {dp = }")

    # 填充動態規劃表格，左上到右下
    for r in range(1, m + 1):
      for c in range(1, n + 1):
        print("-" * 100)
        print(f"{r = }; {c = }")
        print(f"{word1[r - 1] = }; {word2[c - 1] = }")

        # 若兩字元相同，則取左上角的值
        if word1[r - 1] == word2[c - 1]:
          dp[r][c] = dp[r - 1][c - 1]
          print("<<same>> ")
          
        # 若兩字元不同，則取左上、左、上三個方向的最小值加一
        else:
          dp[r][c] = min(dp[r - 1][c - 1], dp[r - 1][c], dp[r][c - 1]) + 1
          print("<<diff>> ")
          print(f"dp[r - 1][c - 1] = {dp[r - 1][c - 1]}; dp[r - 1][c] = {dp[r - 1][c]}; dp[r][c - 1] = {dp[r][c - 1]}")
        print(f"dp[{r}][{c}] = {dp[r][c]}")
        print(f"{dp = }")

    # 回傳動態規劃表格右下角的值，即為最小編輯距離
    return dp[m][n]
```

```python
word1 = "horse"; word2 = "ros" # 3
# word1 = "intention"; word2 = "execution" # 5
Solution().minDistance(word1, word2)
```
