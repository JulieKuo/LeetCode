### 77. Combinations

```python
class Solution:
    def combine(self, n, k):
        # 初始化物件屬性
        self.n = n
        self.k = k
        self.res = []  # 用來存放最終結果的列表

        # 呼叫深度優先搜尋函數，起始索引為0，目前的子集為空列表
        self.dfs(index = 0, subset = [])

        return self.res  # 回傳最終結果

    def dfs(self, index, subset):
        print("-" * 150)
        print(f"index = {index}, subset = {subset}")

        # 如果目前的子集大小等於要達到的組合大小
        if len(subset) == self.k:
            self.res.append(subset)  # 將目前的子集加入最終結果列表
            print(f"-> res = {self.res}")
            return  # 結束當前遞迴

        # 從目前索引開始，遍歷可能的數字
        for i in range(index+1, self.n+1):
            print(f"dfs -> index = {index}, i = {i}, subset = {subset}")
            # 遞迴呼叫，索引更新為下一個數字，子集更新為當前子集加上目前數字
            self.dfs(index = i, subset = (subset + [i]))
```

```python
n = 4; k = 2
Solution().combine(n, k)
```
