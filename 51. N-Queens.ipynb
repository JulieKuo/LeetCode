### 51. N-Queens

```python
class Solution:
    def solveNQueens(self, n: int):
        self.n = n
        self.state = [["."] * n for _ in range(n)]  # 創建一個空的棋盤，以點表示每個方格
        self.res = []
        
        self.visited_cols = set() # 追蹤已經訪問的列

        # 追蹤已訪問的對角線(左上到右下)，記錄行和列的差
        # 同一對角線上 "行和列的差" 具有相同的差異
        # ex: 中 - (2,1) = 2-1 = 1、左上 - (1,0) = 1-0 = 1、右下 - (3,2) = 3-2 = 1 位於同一對角線上
        self.visited_diagonals = set()

        # 追蹤已訪問的反對角線(右上到左下)，記錄行和列的和
        # 同一對角線上 "行和列的和" 具有相同的差異
        # ex: 中 - (3,2) = 3+2 = 5、左下 - (1,4) = 1+4 = 5、右上 - (2,3) = 2+3 = 5 位於同一對角線上
        self.visited_antidiagonals = set()

        self.dfs(row = 0)

        return self.res

    def dfs(self, row):
        print("_" * 150)
        # 如果已是最後一row，將當前state加到res中
        if row == self.n:
            self.res.append(["".join(row) for row in self.state])  # 每row的n個值合成一個字
            print(f"* res = {self.res}")
            return

        for col in range(self.n):
            print("-" * 150)
            print(f"(row, col) = ({row}, {col})")
            diff = row - col
            _sum = row + col

            print(f"-> ((col = {col}) not in (visited_cols = {self.visited_cols})) = {(col not in self.visited_cols)}")
            print(f"-> ((diff = {diff}) not in (visited_diagonals = {self.visited_diagonals})) = {(diff not in self.visited_diagonals)}")
            print(f"-> ((_sum = {_sum}) not in (visited_antidiagonals = {self.visited_antidiagonals})) = {(_sum not in self.visited_antidiagonals)}")
            # 如果當前方格中沒有在 (已訪問的列中) 和 (已訪問的對角線中) 和 (已訪問的反對角線中)
            if not ((col in self.visited_cols) or (diff in self.visited_diagonals) or (_sum in self.visited_antidiagonals)):
                self.visited_cols.add(col)  # 新增進已訪問的列中
                self.visited_diagonals.add(diff)  # 新增進已訪問的對角線中
                self.visited_antidiagonals.add(_sum)  # 新增進已訪問的反對角線中
                self.state[row][col] = 'Q' # 放置皇后

                print(f"({row}, {col}) - add - state = {self.state}")
                print(f"row({row}) - Queen - row add")
                self.dfs(row = row+1)

                # 重置狀態
                print(f"=> ({row}, {col}) - cannot achive row({self.n}) - reset")
                print(f"=> init   - state = {self.state}")
                self.visited_cols.remove(col)
                self.visited_diagonals.remove(diff)
                self.visited_antidiagonals.remove(_sum)
                self.state[row][col] = '.'
                print(f"=> remove - state = {self.state}")

            print(f"row({row}) - No Queen - col add")
```

```python
n = 4
Solution().solveNQueens(n)
```
