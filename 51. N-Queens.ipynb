### 51. N-Queens

- 時間複雜度：$O(n!)$
  - 在第一行，我們有 $n$ 個位置可以選擇；在第二行，由於不能與第一行同列，最多剩下 $n-1$ 個位置；第三行最多剩下 $n-2$ 個位置，以此類推。雖然對角線的限制會進一步剪枝（減少實際嘗試的次數），但在最壞情況下的時間複雜度上限仍然是 $O(n!)$。
- 空間複雜度：$O(2n + n^2)$ -> $O(n^2)$
  -  * 遞迴呼叫堆疊（Call Stack）的最大深度為 $n$（即棋盤的行數）。
  -  用來記錄衝突的集合 cols、diag1、diag2 最多各自儲存 $n$ 或 $2n-1$ 個元素，皆為 $O(n)$ 等級。
  -  儲存棋盤狀態的臨時陣列 board 佔用 $O(n^2)$ 空間。

💡 主對角線（左上到右下 \）  
在數學上，這條線的斜率是 $1$，方程式可以寫成：
$$y = x + k \implies y - x = k$$
如果把 $y$ 換成 row，$x$ 換成 col，公式就變成：
$$\text{row} - \text{col} = k \text{（常數）}$$
我們直接看一個 $4 \times 4$ 棋盤的實際座標：
- 穿過左上角的正中央對角線：(0,0), (1,1), (2,2), (3,3)
  - $0 - 0 = \mathbf{0}$
  - $1 - 1 = \mathbf{0}$
  - $2 - 2 = \mathbf{0}$
  - $3 - 3 = \mathbf{0}$
  - 只要在這條線上的格子，row - col 算出來一律都是 0！
- 往右上一格的平行對角線：(0,1), (1,2), (2,3)
  - $0 - 1 = \mathbf{-1}$
  - $1 - 2 = \mathbf{-1}$
  - $2 - 3 = \mathbf{-1}$
- 往左下一格的平行對角線：(1,0), (2,1), (3,2)
  - $1 - 0 = \mathbf{1}$
  - $2 - 1 = \mathbf{1}$
  - $3 - 2 = \mathbf{1}$
結論： 每一條「左上到右下」的斜線，都有一個獨一無二的 row - col 身份證字號。只要我們把這個數字丟進 diag1 集合，下次只要算一下新位置的 row - col 是不是這個數字，就能一秒知道有沒有人在同一條線上了！

💡 副對角線（右上到左下 /）  
在數學上，這條線的斜率是 $-1$，方程式可以寫成：
$$y = -x + k \implies y + x = k$$
換成棋盤座標就是：
$$\text{row} + \text{col} = k \text{（常數）}$$
我們同樣用 $4 \times 4$ 棋盤來看實際座標：
- 穿過中間的副對角線：(0,3), (1,2), (2,1), (3,0)
  - $0 + 3 = \mathbf{3}$
  - $1 + 2 = \mathbf{3}$
  - $2 + 1 = \mathbf{3}$
  - $3 + 0 = \mathbf{3}$
  - 在這條線上的格子，row + col 算出來一律都是 3！
- 左上角最邊邊的那格：(0,0)
  - $0 + 0 = \mathbf{0}$ （自成一條長度為 1 的對角線）
結論： 每一條「右上到左下」的斜線，她們的身份證字號就是 row + col。

```python
class Solution:
    def solveNQueens(self, n: int) -> list[list[str]]:
        # 儲存所有合法棋盤配置的結果陣列
        res = []
        
        # 初始化棋盤，先全部填入 '.'
        board = [["."] * n for _ in range(n)]
        
        # 用於快速檢查衝突的集合（空間換時間）
        cols = set()        # 記錄哪些「列」已有皇后
        diag1 = set()       # 記錄哪些「主對角線 (\)」已有皇后 (特徵值: row - col)
        diag2 = set()       # 記錄哪些「副對角線 (/)」已有皇后 (特徵值: row + col)
        
        def backtrack(row: int):
            # 成功放置了 n 個皇后，找到一組解
            if row == n:
                # 將棋盤的每一行從陣列轉換為字串，並加入結果中
                res.append(["".join(r) for r in board])
                return
            
            # 嘗試在當前行的每一個列位置放置皇后
            for col in range(n):
                # 檢查當前位置是否會被其他皇后攻擊
                if col in cols or (row - col) in diag1 or (row + col) in diag2:
                    continue  # 若有衝突，直接跳過此列
                
                # 做出選擇：放置皇后並記錄狀態
                board[row][col] = "Q"
                cols.add(col)
                diag1.add(row - col)
                diag2.add(row + col)
                
                # 遞迴前往下一行
                backtrack(row + 1)
                
                # 回溯：撤銷剛才的選擇，恢復狀態
                board[row][col] = "."
                cols.remove(col)
                diag1.remove(row - col)
                diag2.remove(row + col)
                
        # 從第 0 行開始進行回溯搜索
        backtrack(0)
        return res
```

```python
n = 4
Solution().solveNQueens(n)
```
