### 438. Find All Anagrams in a String (Substrings and Anagrams)

#### Question
Given two strings, string_a and string_b, determine whether any anagram of string_b, occurs as a substring of string_a.

> [!NOTE]
> 在 string_a 中，是否存在一個子字串（Substring），它與 string_b 互為字母異位詞（Anagram）？ Sub string_a 的長度會等於 string_b。

#### Clarifying Questions
- What is an anagram?
  - Two strings are anagrams of each other if you can rearrange the characters of one string to form the other. For this problem, all the letters must be used, and punctuation and whitespace are ignored.
- Does case matter when comparing substrings to find an anagram?
  - No, your code should ignore case when comparing.
- Are whitespace characters included when finding an anagram?
  - No, whitespace characters in either string are ignored.
- Are punctutation characters included when finding an anagram?
  - No, punctutation characters in either string are ignored.
- Can the entire string_a be considered when trying to find an anagram match?
  - The substring of string_a that matches the anagram of string_b must be from consecutive characters (ignoring punctuation and whitespace) in string_a.
- What if string_a or string_b is empty?
  - You can assume that both string_a and string_b are nonempty and contain at least one letter each.
  
#### Examples
| string_a | string_b | Result |
| :--- | :--- | :--- |
| Here come dots (中) and lines that helped build America | The  (中) Morse Code | TRUE |
| crowd | word | TRUE |
| There is no anagram here | some word | FALSE |

**時間複雜度: $O(2n + 3m) -> O(n + m)$**  
**空間複雜度: $O(n + 3m) -> O(n + m)$**

- Leetcode 的 s and p 只會由英文字母組成；Google 的題目包還標點符號。Leetcode 要判斷 anagrams 的起始位置。Google 的題目只要判斷有沒有 anagrams。結合兩題需求，把題目想成是要判斷 anagrams 的起始位置，但是位置已清洗過標點符號後的 string 來做計算。

```python
class Solution:
    def findAnagrams(self, s: str, p: str) -> list[int]:
        # 1. 資料清洗
        s_clean = [c.lower() for c in s if c.isalnum()] # time, space: O(n)
        p_clean = [c.lower() for c in p if c.isalnum()] # time, space: O(m)

        s_length, p_length = len(s_clean), len(p_clean)
        
        if s_length < p_length: 
            return []

        # 2. 初始化頻率表
        target_counts = {} # space: O(m)
        for char in p_clean: # time: O(m)
            target_counts[char] = target_counts.get(char, 0) + 1
        
        result = []
        window_counts = {} # space: O(m)
        left = 0
        
        # 3. 滑動視窗
        for right in range(s_length): # time: O(n)
            char_right = s_clean[right]
            window_counts[char_right] = window_counts.get(char_right, 0) + 1
            
            # 當視窗長度超過 m，移除左側字元
            if right - left + 1 > p_length:
                char_left = s_clean[left]
                window_counts[char_left] -= 1
                if window_counts[char_left] == 0:
                    del window_counts[char_left]
                left += 1
            
            # 4. 比較字典是否相等
            if window_counts == target_counts: # time: O(m)
                result.append(left)
                
        return result
```

```python
s = "Here come dots (中) and lines that helped build America"
p = "The (中) Morse Code"

s = "cbabcebabacd"
p = "abc"
Solution().findAnagrams(s, p)
```

#### List

**時間複雜度: $O(2n + 2m) -> O(n + m)$**  
**空間複雜度: $O(n + m) -> O(n + m)$**

- 只有在確定只會出現英文的情況下才能使用，有數字和中文的話我們沒辦法確定索引。只適用於 Leetcode 383. Ransom Note
- 空間更省：字典（dict）需要額外儲存 Hash Key 和處理碰撞（Collision）的空間；而陣列（list）只佔用連續的 26 個整數空間。
- 速度更快：ord(char) - ord("a") 的數學運算通常比字典的雜湊函數運算更快。

```python
class Solution:
    def findAnagrams(self, s: str, p: str) -> list[int]:
        # 1. 資料清洗
        s_clean = [c.lower() for c in s if c.isalpha()] # time, space: O(n)
        p_clean = [c.lower() for c in p if c.isalpha()] # time, space: O(m)

        s_length, p_length = len(s_clean), len(p_clean)
        
        if s_length < p_length: 
            return []

        # 2. 初始化 26 字母桶子
        target_counts = [0] * 26 # space: O(1)
        window_counts = [0] * 26 # space: O(1)
        
        # 統計 p 的頻率
        for char in p_clean: # time: O(m)
            target_counts[ord(char) - ord('a')] += 1
            
        result = []
        left = 0
        # 3. 滑動視窗
        for right in range(s_length): # time: O(n)
            # 右字元入窗
            window_counts[ord(s_clean[right]) - ord('a')] += 1
            
            # 若視窗太長，左字元出窗
            if right - left + 1 > p_length:
                window_counts[ord(s_clean[left]) - ord('a')] -= 1
                left += 1
                
            # 4. 直接比對兩個陣列
            if window_counts == target_counts: # time: O(26) -> O(1)
                result.append(left)
                
        return result
```

```python
s = "Here come dots and lines that helped build America"
p = "The Morse Code"

s = "cbabcebabacd"
p = "abc"
Solution().findAnagrams(s, p)
```
