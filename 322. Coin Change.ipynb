### 322. Coin Change

**時間複雜度: O($amount \times n$)**  
**空間複雜度: O($amount$)**

$n$: 可用硬幣數

```python
from typing import List

class Solution:
    def coinChange(self, coins: List[int], amount: int) -> int:
        # dp[i]表示湊齊金額i所需的最少硬幣數量，每個元素初始值為正無窮大
        dp = [float("inf")] * (amount + 1)
        dp[0] = 0 # 金額0需0個硬幣
        print(f"{dp = }")

        # 從0到amount依次計算最少硬幣數量
        for i in range(1, (amount + 1)):
            print("_" * 500)
            for coin in coins:
                print("-" * 500)
                print(f"{i = }, {coin = }, {(i - coin) = }")
                # 檢查當前金額減去當前硬幣面值是否大於等於0
                if (i - coin) >= 0:
                    print(f"-> {dp[i] = }, {(dp[i - coin] + 1) = }")

                    # 比較 "金額i所需的最少硬幣數量" 和 "金額(i-coin)所需的最少硬幣數量 + 1個coin"，取最小值更新dp[i]
                    dp[i] = min(dp[i], dp[i - coin] + 1)

                    print(f"--> {dp[i] = }, {dp = }")

        if (dp[amount] == float("inf")):
            # 如果dp[amount]仍為正無窮大，表示無法湊齊該金額，返回-1
            return -1
        else:
            # 否則返回dp[amount]表示的最少硬幣數量
            return dp[amount]
```

```python
coins = [1,2,5]
amount = 11 # 3
Solution().coinChange(coins, amount)
```
