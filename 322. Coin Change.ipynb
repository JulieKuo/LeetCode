### 322. Coin Change

**時間複雜度: O($amount \times n$)**  
**空間複雜度: O($amount$)**

$n$: 可用硬幣數

```python
from typing import List

class Solution:
    def coinChange(self, coins: List[int], amount: int) -> int:
        # dp[i] 代表湊出金額i所需的最少硬幣數
        # 初始化dp陣列,除了dp[0]=0外,其餘設為無限大
        dp = [float("inf")] * (amount + 1)
        dp[0] = 0

        # 對每個硬幣面額進行迭代
        for coin in coins:
            # 從coin開始迭代到amount,因為金額小於coin的無法用當前硬幣湊出
            for i in range(coin, amount + 1):
                # dp[i-coin] + 1 表示使用當前硬幣後所需的硬幣數
                # # 比較 "金額i所需的最少硬幣數量" 和 "金額(i-coin)所需的最少硬幣數量 + 1個coin"，取最小值更新dp[i]
                dp[i] = min(dp[i - coin] + 1, dp[i])
                
                print(f"{coin=}, {i=}")
                print(f"{(dp[i - coin] + 1)=}, {dp[i]=}")
                print(f"{dp=}\n")

        # 如果dp[amount]仍為無限大,表示無法湊出該金額,回傳-1
        # 否則回傳所需最少硬幣數
        return dp[amount] if dp[amount] != float("inf") else -1
```

```python
coins = [1,2,5]
amount = 11 # 3
Solution().coinChange(coins, amount)
```
