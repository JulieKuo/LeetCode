### 399. Evaluate Division

```python
from typing import List

class Solution:
    def calcEquation(self, equations: List[List[str]], values: List[float], queries: List[List[str]]) -> List[float]:
        # 建立有向圖，graph為圖的表示，equations中的方程式和values中的值用來建立邊
        graph = self.buildGraph(equations, values)
        final_answer = []  # 存放最終結果的列表

        print(f"graph = {graph}")
        print(f"queries = {queries}")
        # 對於每一個查詢進行處理
        for query in queries:
            dividend, divisor = query # 被除數, 除數
            print("_" * 150)
            print(f"dividend = {dividend} , divisor = {divisor}")

            # 如果被除數或除數不在圖中，則結果為-1.0
            if (dividend not in graph) or (divisor not in graph):
                final_answer.append(-1.0)
            else:
                visited = set()  # 記錄訪問過的節點
                answer = [-1.0]  # 存放結果的列表 (修改列表中的元素，而不是直接修改變數的值。即傳遞可變對象(如列表)而不是不可變對象(如數字))
                temp = 1.0  # 紀錄當前的乘積
                self.dfs(node = dividend, target = divisor, graph = graph, visited = visited, answer = answer, temp = temp)
                final_answer.append(answer[0])  # 添加結果到最終答案列表

        return final_answer


    def buildGraph(self, equations: List[List[str]], values: List[float]) -> dict:
        graph = {}  # 圖的表示，使用字典來表示有向邊

        # 遍歷方程式和值的列表
        for i in range(len(equations)):
            dividend, divisor = equations[i] # 被除數, 除數
            value = values[i]

            # 如果被除數不在圖中，則添加到圖中
            if dividend not in graph:
                graph[dividend] = {}

            # 如果除數不在圖中，則添加到圖中
            if divisor not in graph:
                graph[divisor] = {}

            # 添加有向邊和邊的值
            graph[dividend][divisor] = value
            graph[divisor][dividend] = 1 / value

        return graph
    

    def dfs(self, node: str, target: str, graph: dict, visited: set, answer: List[float], temp: float) -> None:
        print("-" * 150)
        print(f"node = {node}, target = {target}, temp = {temp}, answer = {answer}, graph = {graph}, visited = {visited}")
        # 如果節點已經訪問過，則返回
        if node in visited:
            print(f"(node = {node}) in (visited = {visited})")
            return

        visited.add(node)  # 將節點標記為已訪問       
        print(f"visited = {visited}")
        
        if node == target:
            answer[0] = temp  # 如果節點是目標節點，則更新結果
            print(f"(node = {node}) == (target = {target})")
            print(f"answer = {answer}")
            return

        # 對於節點的每個鄰接節點，執行深度優先搜索
        for neighbor, value in graph[node].items():
            print(f"{node} - {neighbor} - dfs")
            '''
            a/b = 2
            b/c = 4
            a/c = a/b * b/c = 2 * 4 = 8
            '''
            self.dfs(node = neighbor, target = target, graph = graph, visited = visited, answer = answer, temp = (temp * value))
            if neighbor == target:
                break
```

```python
equations = [["a", "b"],["b", "c"], ["a", "e"]]
values = [2, 4, 5]
queries = [["a","c"],["b","a"],["a","x"],["a","e"],["a","a"]]
```

```python
Solution().calcEquation(equations, values, queries)
```
