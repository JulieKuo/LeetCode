### 1376. Time Needed to Inform All Employees

```python
n = 6
headID = 5
manager =    [1,3,3,5,2,-1]
informTime = [0,3,3,2,0, 1]
# 6
```

```python
class Solution:
    def numOfMinutes(self, n, headID, manager, informTime):
        # 定義一個遞迴函數 dfs，用於計算從員工 i 開始傳遞訊息所需的時間
        def dfs(i):
            print("_" * 150)
            print(f"i = {i}")

            # 如果員工 i 有直屬上司
            if manager[i] != -1:
                print(f"(manager[i] = manager[{i}] = {manager[i]}) != -1")
                print(f"-> init: informTime[i]  = informTime[{i}] = {informTime[i]}")

                # 遞迴呼叫 dfs 函數，計算從直屬上司開始的累加傳遞時間，並加到 informTime[i] 中
                add = dfs(manager[i])
                informTime[i] += add
                print(f"-> add = {add}")
                print(f"-> new:  informTime[i]  = informTime[{i}] = {informTime[i]}")

                # 將直屬上司標記為已訪問，避免重複累加
                manager[i] = -1

                print(f"-> manager[i] = manager[{i}] = {manager[i]}")
                print(f"manager = {manager}")

            # 回傳從員工 i 開始的累計總傳遞時間
            print(f"informTime = {informTime}")
            print(f"return (informTime[{i}] = {informTime[i]})")
            print("-" * 150)
            return informTime[i]

        # 使用 map 函數對所有員工調用 find 函數，並取得傳遞時間的最大值
        return max(map(dfs, range(n)))
        return max([dfs(i) for i in range(n)]) # 相同功能
```

```python
Solution().numOfMinutes(n, headID, manager, informTime)
```
