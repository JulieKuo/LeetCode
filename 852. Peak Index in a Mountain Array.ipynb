## 852. Peak Index in a Mountain Array (Up and Down Array)

#### Question
You are asked to write a function to find the index of the maximum value of an array of data, under the assumption that the first part of the array is strictly increasing, and then it becomes strictly decreasing.

#### Example
If the students asks for an example, offer these:

If we are given this array:

```
[5, 12, 17, 22, 25, "37", 23, 11, 6, 4, 2, 1]
```

our code should return 5, the index of number 37 in the array.  
(hint: what is the index of the first element of an array?)

The same example, here for easier copy/paste:

```
sorted_array = [5, 12, 17, 22, 25, 37, 23, 11, 6, 4, 2, 1]
result = 5
```

#### Clarifying Questions
Candidates may ask some or all of the following questions to clarify:
- What sould be returned for an empty array?
  - Return -1 in the case of the empty input array
- Are the array values all integers?
  - Yes, the array contains only integer values
- Can values in the array be repeated?
  - Consecutive values in the array will not repeat. It is possible that the same value appears in both the increasing part and the decreasing part of the array values.

- 時間複雜度：$O(\log N)$
  - 每次迴圈我們都會將搜尋區間砍半。若陣列長度為 $N$，則最多經過 $\log_2 N$ 次比較就能找到結果。對於 100 萬筆資料，僅需約 20 次比較即可找到峰值。
- 空間複雜度：$O(1)$
  - 僅使用 left, right, mid 三個變數，不隨輸入資料量增加而增加空間消耗。

````python
from typing import List

class Solution:
    def peakIndexInMountainArray(self, arr: List[int]) -> int:
        # 處理邊界情況：若陣列為空，回傳 -1
        if not arr:
            return -1
        
        # 初始化左右指標
        left, right = 0, len(arr) - 1
        
        # 當左右指標尚未重合時持續搜尋
        while left < right:
            # 取得中間索引，使用 (left + right) // 2 避免溢位（在 Python 中較無此問題，但為好習慣）
            mid = (left + right) // 2
        
            # 比較中間值與下一個值
            if arr[mid] < arr[mid + 1]:
                # 代表還在爬坡中，峰值在右邊，移動 left 指標
                left = mid + 1
            else:
                # 代表正在下坡或就在峰值，峰值可能就是 mid 或在左邊
                right = mid
            
        # 最後 left 與 right 會重合在峰值索引處
        return left
````

````python
arr = [5, 12, 17, 22, 25, 37, 23, 11, 6, 4, 2, 1]
Solution().peakIndexInMountainArray(arr)
````
